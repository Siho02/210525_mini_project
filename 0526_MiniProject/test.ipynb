```python
'''practice of class and 상속
'''

class Iphone:
    ''' Iphone have name, nuber, version(optional)
    three method : update_version, calling, search_contact
    '''
    def __init__(self, name, number,version=1.0):
        # we will add name, number, version(optional)
        self.name = name
        self.number = number
        self.version = version
        print("개통을 축하합니다. %s님" % name)

    def update_version(self):
        '''update_version will add 0.1 to previous version and print it'''
        self.version += 0.1
        print("os version is updated!")

    def calling(self):
        '''calling method will input a phone number and print the number + calling string'''
        call_num = input("phone number : ")
        print(call_num+' calling~')

    def contact(self):
        '''contact will read a contact.txt and search the name and number'''
        with open('dataSet/contact.txt', 'r', encoding='utf-8') as contact_list:
            datas = contact_list.readlines()

        lst = []
        for data in datas:
            lst.append(data.split(' '))
        string = input()
        for idx in range(len(lst)):
            for idx2 in range(2):
                if string in lst[idx][idx2]:
                    print(lst[idx])
                else:
                    if idx == len(lst)-1 and idx2 == 1:
                        print("NO search result")

class Iphone2(Iphone):
    ''' Iphone2 is inherited by Iphone and have name, nuber, version(optional)
    additional method : message
    '''
    def __init__(self, name, number, version=2.0):
        super().__init__(name, number, version)

    def send_message(self):
        ''' message method take a msg & number by string and print msg
        msg is too short(length = 0) will print a failure message
        '''
        msg = input("text : ")
        if len(msg) < 1:
            print("FAILED : text is too short")
        else:
            num = input("phone number : ")
            print("Message was sented to user"+num)

#Iphone3 is sub class of Iphone2
class Iphone3(Iphone2):
    '''Iphone3 is inherited by Iphone2 and have name, nuber, version(optional)
    additional method : wifi
    '''
    def __init__(self, name, number, version=3.0):
        super().__init__(name, number, version)

    def wifi_check(self):
        '''wifi method take password. if password is right we will
        print connected else print connect fail'''
        wifi_pwd = '1234'
        pwd = input("enter the password : ")

        if pwd == wifi_pwd:
            print("conneted")
        else:
            print("connet fail")
```
